```python
files = [
    {
        "var_src_folder": "netflix_directors",
        "var_trg_folder": "netflix_directors"
    },
    {
        "var_src_folder": "netflix_cast",
        "var_trg_folder": "netflix_cast"
    },
    {
        "var_src_folder": "netflix_countries",
        "var_trg_folder": "netflix_countries"
    },
    {
        "var_src_folder": "netflix_category",
        "var_trg_folder": "netflix_category"
    }
]

print(files)

```

```python
for file in files:
    
    src = file["var_src_folder"]
    trg = file["var_trg_folder"]
    
    print(f"Processing {src}")
    
    df = spark.read.table(f"dbo.{src}")
    
    df.write.mode("overwrite").format("delta").saveAsTable(f"dbo.{trg}")
 
```
